```python
from bokeh.models.widgets import Panel, Tabs
from bokeh.io import output_notebook, show
from bokeh.plotting import figure

output_notebook()
```

```python
fruits = ['Apples', 'Pears', 'Nectarines', 'Plums', 'Grapes', 'Strawberries']

plot1 = figure(x_range=fruits, plot_height=300, toolbar_location=None, title="Fruit Counts")
plot1.vbar(x=fruits, top=[5, 3, 4, 2, 4, 6], width=0.9)
plot1.xgrid.grid_line_color = None
plot1.y_range.start = 0
#show(plot1)
```

```python
from bokeh.models import ColumnDataSource
from bokeh.core.properties import value

fruits = ['Apples', 'Pears', 'Nectarines', 'Plums', 'Grapes', 'Strawberries']
years = ["2015", "2016", "2017"]
colors = ["#c9d9d3", "#718dbf", "#e84d60"]

data = {'fruits' : fruits,
        '2015'   : [2, 1, 4, 3, 2, 4],
        '2016'   : [5, 3, 4, 2, 4, 6],
        '2017'   : [3, 2, 4, 4, 5, 3]}

source = ColumnDataSource(data=data)

plot2 = figure(x_range=fruits,  plot_height=300, title="Fruit Counts by Year", toolbar_location=None, tools="")

plot2.vbar_stack(years, x='fruits', width=0.9, color=colors, source=source, legend=[value(x) for x in years])

plot2.y_range.start = 0
plot2.x_range.range_padding = 0.1
plot2.xgrid.grid_line_color = None
plot2.axis.minor_tick_line_color = None
plot2.outline_line_color = None
plot2.legend.location = "top_left"
plot2.legend.orientation = "horizontal"

#show(plot2)
```

```python
tab1 = Panel(child=plot1, title="Bar Plot")
tab2 = Panel(child=plot2, title="StackedBar Plot")

tabs = Tabs(tabs=[ tab1, tab2 ])
show(tabs)
```
